```python
#%pip install kagglehub
import kagglehub

# Download latest version
path = kagglehub.dataset_download("Cornell-University/arxiv")

print("Path to dataset files:", path)
```

```python
import json

subset = []
max_docs = 25000  # Aproximadamente 1% del total

with open(path + "/arxiv-metadata-oai-snapshot.json", 'r') as f:
    for i, line in enumerate(f):
        if i >= max_docs:
            break
        try:
            record = json.loads(line)
            subset.append({
                "id": record.get("id"),
                "title": record.get("title", ""),
                "abstract": record.get("abstract", ""),
                "authors": record.get("authors", ""),
                "categories": record.get("categories", ""),
                "update_date": record.get("update_date", "")
            })
        except json.JSONDecodeError:
            continue

# Guardar el subset como nuevo archivo corpus.json
with open("corpus.json", "w") as out_file:
    json.dump(subset, out_file)
```

```python
import json
import pandas as pd

# Cargar corpus desde el archivo
with open("corpus.json", "r") as f:
    raw_corpus = json.load(f)

# Convertir a DataFrame
df_corpus = pd.DataFrame(raw_corpus)

# Ver columnas disponibles
print(df_corpus.columns)
df_corpus.head()
```

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words("english"))

def preprocess_text(text):
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return " ".join(tokens)

# Crear columna "processed_text" combinando tÃ­tulo y resumen
df_corpus["text"] = df_corpus["title"].fillna('') + " " + df_corpus["abstract"].fillna('')
df_corpus["processed_text"] = df_corpus["text"].apply(preprocess_text)
```

TFâ€“IDF y BM25

```python
#%pip install scikit-learn rank_bm25 nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from rank_bm25 import BM25Okapi
from nltk.tokenize import word_tokenize
import numpy as np

# Extraemos los textos preprocesados
corpus_texts = df_corpus["processed_text"].tolist()

# --------------------
# TFâ€“IDF Index
# --------------------
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus_texts)

# --------------------
# BM25 Index
# --------------------
# Tokenizamos cada documento para BM25
tokenized_corpus = [word_tokenize(text) for text in corpus_texts]
bm25_model = BM25Okapi(tokenized_corpus)
```

Funciones de BÃºsqueda

```python
# --------------------
# FunciÃ³n TFâ€“IDF
# --------------------
def search_tfidf(query, top_k=10):
    query_vec = tfidf_vectorizer.transform([query])
    scores = np.dot(tfidf_matrix, query_vec.T).toarray().ravel()
    top_indices = np.argsort(scores)[::-1][:top_k]

    return [
        {
            "id": df_corpus.iloc[i]["id"],
            "title": df_corpus.iloc[i]["title"],
            "abstract": df_corpus.iloc[i]["abstract"][:300] + "...",
            "score": scores[i]
        }
        for i in top_indices
    ]


# --------------------
# FunciÃ³n BM25
# --------------------
def search_bm25(query, top_k=10):
    tokenized_query = word_tokenize(query.lower())
    scores = bm25_model.get_scores(tokenized_query)
    top_indices = np.argsort(scores)[::-1][:top_k]

    return [
        {
            "id": df_corpus.iloc[i]["id"],
            "title": df_corpus.iloc[i]["title"],
            "abstract": df_corpus.iloc[i]["abstract"][:300] + "...",
            "score": scores[i]
        }
        for i in top_indices
    ]
```

```python
from sentence_transformers import SentenceTransformer
import faiss

# Instanciar modelo de embeddings
embed_model = SentenceTransformer('all-MiniLM-L6-v2')  # 384 dimensiones

# Obtener los embeddings para los documentos
doc_embeddings = embed_model.encode(corpus_texts, show_progress_bar=True)

# Crear Ã­ndice FAISS
dim = doc_embeddings.shape[1]
index_faiss = faiss.IndexFlatL2(dim)
index_faiss.add(doc_embeddings)
```

```python
def search_faiss(query, top_k=10):
    query_vec = embed_model.encode([query])
    distances, indices = index_faiss.search(query_vec, top_k)

    return [
        {
            "id": df_corpus.iloc[i]["id"],
            "title": df_corpus.iloc[i]["title"],
            "abstract": df_corpus.iloc[i]["abstract"][:300] + "...",
            "score": float(distances[0][rank])
        }
        for rank, i in enumerate(indices[0])
    ]
```

```python
# Leer las queries desde el archivo
with open("queries.txt", "r") as f:
    queries = [line.strip() for line in f if line.strip()]

# Ejecutar y guardar resultados
results_all = {}

for query in queries:
    results_all[query] = {
        "tfidf": search_tfidf(query),
        "bm25": search_bm25(query),
        "faiss": search_faiss(query)
    }

# Ejemplo de visualizaciÃ³n de resultados
for query in queries:
    print(f"\n QUERY: {query}")
    for model in ["tfidf", "bm25", "faiss"]:
        print(f"\nðŸ”¹ {model.upper()}")
        for r in results_all[query][model][:3]:
            print(f"- {r['id']}: {r['title']}")
```

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

# Cargar variables de entorno
load_dotenv()

# Leer la clave desde el archivo .env
api_key = os.getenv("GEMINI_API_KEY")

# Configurar Gemini
genai.configure(api_key=api_key)

```

```python

model = genai.GenerativeModel("gemini-2.5-flash")

def generate_rag_answer_gemini(query):
    top_docs = search_faiss(query, top_k=3)
    
    context = "\n\n".join([
        f"Title: {doc['title']}\nAbstract: {doc['abstract']}" for doc in top_docs
    ])

    prompt = f"""Query: "{query}"\n\n{context}\n\nAnswer:"""

    response = model.generate_content(prompt)

    return {
        "query": query,
        "context_docs": top_docs,
        "answer": response.text
    }
```

```python
rag_results_gemini = {}

for query in queries:
    print(f" Generando respuesta con Gemini para: {query}")
    rag_results_gemini[query] = generate_rag_answer_gemini(query)
```

Comparacion respuesta RAG

```python
import re

def evaluar_coherencia_rag(rag_result, min_porcentaje=15):
    # 1. Respuesta generada por Gemini
    respuesta = rag_result["answer"].lower()

    # 2. Concatenar todos los resÃºmenes usados como contexto
    contexto = " ".join([doc["abstract"].lower() for doc in rag_result["context_docs"]])

    # 3. Extraer palabras clave significativas (>=5 letras, sin stopwords comunes)
    palabras_respuesta = re.findall(r'\b\w{5,}\b', respuesta)
    palabras_contexto = re.findall(r'\b\w{5,}\b', contexto)

    # Convertir a sets para comparar
    palabras_unicas = set(palabras_respuesta)
    coincidencias = palabras_unicas & set(palabras_contexto)

    porcentaje = (len(coincidencias) / len(palabras_unicas)) * 100 if palabras_unicas else 0

    # 4. EvaluaciÃ³n bÃ¡sica
    coherente = porcentaje >= min_porcentaje

    return {
        "total_palabras_clave": len(palabras_unicas),
        "coincidencias": len(coincidencias),
        "porcentaje": round(porcentaje, 2),
        "coherente": " SÃ­" if coherente else " No"
    }
```

```python
for query in queries:
    resultado = evaluar_coherencia_rag(rag_results_gemini[query])
    print(f" {query}")
    print(f"  - Palabras clave: {resultado['total_palabras_clave']}")
    print(f"  - Coincidencias con contexto: {resultado['coincidencias']}")
    print(f"  - Porcentaje: {resultado['porcentaje']}%")
    print(f"  - Â¿Coherente? {resultado['coherente']}\n")



```

EvaluaciÃ³n Comparativa

```python
def compare_models(query, results_dict):
    tfidf_ids = [doc["id"] for doc in results_dict[query]["tfidf"]]
    bm25_ids  = [doc["id"] for doc in results_dict[query]["bm25"]]
    faiss_ids = [doc["id"] for doc in results_dict[query]["faiss"]]

    common_tfidf_bm25 = set(tfidf_ids) & set(bm25_ids)
    common_tfidf_faiss = set(tfidf_ids) & set(faiss_ids)
    common_bm25_faiss = set(bm25_ids) & set(faiss_ids)

    all_three = set(tfidf_ids) & set(bm25_ids) & set(faiss_ids)

    return {
        "query": query,
        "TFIDF-BM25": len(common_tfidf_bm25),
        "TFIDF-FAISS": len(common_tfidf_faiss),
        "BM25-FAISS": len(common_bm25_faiss),
        "All_3": len(all_three)
    }
```

```python
import pandas as pd

comparison_table = []
for q in queries:
    comparison_table.append(compare_models(q, results_all))

df_comparison = pd.DataFrame(comparison_table)
```
