```python
import math
import string

# Leer archivo
with open("01_corpus_turismo.txt", "r", encoding="utf-8") as file:
    documentos = [line.strip().lower() for line in file]

# Función para limpiar texto
def limpiar_texto(texto):
    texto = texto.translate(str.maketrans('', '', string.punctuation))
    return texto.split()

# Tokenizar
documentos_tokenizados = [limpiar_texto(doc) for doc in documentos]

# Vocabulario
vocabulario = sorted(set(palabra for doc in documentos_tokenizados for palabra in doc))
N = len(documentos_tokenizados)

# Construcción de matriz
matriz = []
for termino in vocabulario:
    fila = [termino]
    df_t = 0
    for doc in documentos_tokenizados:
        freq = doc.count(termino)
        if freq > 0:
            df_t += 1
        fila.append(freq)
    idf = math.log10(N / df_t) if df_t != 0 else 0
    fila.append(round(idf, 4))
    matriz.append(fila)

# Formato mejorado para impresión
col_width = 12  # ancho fijo para cada columna
header = ["Término"] + [f"Doc {i+1}" for i in range(N)] + ["IDF"]
print("".join(word.ljust(col_width) for word in header))
print("-" * col_width * len(header))

for fila in matriz:
    print("".join(str(item).ljust(col_width) for item in fila))

print(f"\nNúmero total de términos únicos en el vocabulario: {len(vocabulario)}")
```
